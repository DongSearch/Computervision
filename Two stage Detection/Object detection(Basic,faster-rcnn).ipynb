```python
# box :[x1,y1,x2,y2]
def compute_iou(boxA,boxB):
  # the point at the left up corner
  xA = max(boxA[0],boxB[0])
  yA = max(boxA[1],boxB[1])
  # the point at the right bottom corner
  xB = min(boxA[2],boxB[2])
  yB = min(boxA[3],boxB[3])

  intersection = max(xB-xA,0) * max(yB-yA,0) # area of intersection
  areaA = (boxA[2]-boxA[0]) *(boxA[3]-boxA[1]) # area of boxA
  areaB = (boxB[2]-boxB[0]) *(boxB[3]-boxB[1]) # area of boxA
  union = areaA + areaB - intersection

  return intesection / union if union > 0 else 0


```

```python
from torchvision.models.detection import fasterrcnn_resnet50_fpn, FasterRCNN_ResNet50_FPN_Weights
weights = FasterRCNN_ResNet50_FPN_Weights.DEFAULT # default of MSCOCO
det_model = fasterrcnn_resnet50_fpn(weights=weights).to(device)
det_model.eval()
preproc = weights.transforms() # automatically transform fitting to COCO dataset
```

```python
import urllib.request
from google.colab.patches import cv2_imshow

url = "http://images.cocodataset.org/val2017/000000039769.jpg"
urllib.request.urlretrieve(url,"test_img.jpg")
img = Image.open("test_img.jpg").convert("RGB")
canvas_bgr = cv2.CvtColor(np.array(img), cv2.COLOR_RGB2BGR)
cv2_imshow(canvas_bgr)
print(img.size)#image size
```

```python
x = preproc(img).unsqueeze(0).to(device) # do preprocessing and add batch dimension
with torch.no_grad():
  output = det_model(x) # return dictionary including boxes, labels, scores
  out = det_model(x)[0] # take inforamtion of first image
print(output)
print()
print(out)

boxes = out['boxes'].cpu().numpy()
labels = out['labels'].cpu().numpy()
scores= out['boxes'].cpu().numpy()

# set threshold
thr = 0.5
keep = scores >= thr
boxes,labels,scores = boxes[keep],labels[keep], scores[keep]

visual = np.array(img).copy()

for (x1,y1,x2,y2), s, lb in zip(boxes,scores,labels) :
  cv2.rectangle(visual,(int(x1),int(y1)),(int(x2),int(y2)),(0,255,0),2)
  cv2.putText(visual,f"id:{int(lb)} {s:.2f}",(int(x1),int(y1)-5), cv2.FONT_HERSHEY_SIMPLEX.0.6,(0,0,255),2)

visual_rgb = cv2.cvtColor(visual, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,8))
plt.imshow(vis_rgb)
plt.title('Faster-RCNN Prediction')
plt.axis('off')
plt.show()
```

```python
# define iou using pytorch
#iou =[x,y,w,h]
import torch

def iou_pytorch(box1,box2):
  x1 = torch.max(box1[...,0],box2[...,0])
  y1 = torch.max(box1[...,1],box2[...,1])
  x2 = torch.max(box1[...,2],box2[...,2])
  y2 = torch.max(box1[...,3],box2[...,3])

  inter_w = torch.clamp(x2 - x1, min = 0)
  inter_h = torch.clamp(y2 - y1, min = 0)
  inter = inter_w * inter_h
  area1 = (box1[...,2]-box1[...,0])*(box1[...,3]-box1[...,1])
  area2 = (box2[...,2]-box2[...,0])*(box2[...,3]-box2[...,1])

  union = area1 + area2 - inter
  iou = inter/(union+1e-7)
  return iou

```
